```python
from sentence_transformers import SentenceTransformer
from sklearn.cluster import KMeans
```

```python
productos = df['PRODUCTO_SERVICIO']
```

```python
# Dataset
df1 = pd.DataFrame({'PRODUCTO_SERVICIO':productos})
df1
```

```python
from sentence_transformers import SentenceTransformer
sentences = df1['PRODUCTO_SERVICIO'].to_list()

# Este es un modelo que genera la representación vectorial de las descripciones de cada producto, este modelo fue
# entrenado sobre un dataset en español, por lo que deberíamos esperar un buen rendimiento
model_name = 'hiiamsid/sentence_similarity_spanish_es'
model = SentenceTransformer(model_name)

embeddings = model.encode(sentences)
print(embeddings)
```

```python
embeddings
```

```python
# Eligiendo el numero de clusters con el método Elbow
distortions = []
K = (5,10,15,20,25,30,35,45,50,55,60)
for k in K:
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(embeddings)
    distortions.append(kmeans.inertia_)

plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```

```python
kmeans_model = KMeans(n_clusters=30)
kmeans_model.fit(embeddings)
```

```python
df1['cluster'] = kmeans_model.labels_
```
